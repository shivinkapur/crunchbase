```python
%%R

investments= read.csv('Data/investments.csv')
dim(investments)
```

```python
%%R

#colnames(investments)
summary(investments)

investments=data.frame(investments)
summary(investments)
```

```python
%%R

investments$raised_amount_usd= as.numeric(investments$raised_amount_usd)
```

```python
%%R -w 1000 -h 3000

#Amount invested in a company
funds_per_company = ddply(investments, .(company_name), summarize, total = sum(raised_amount_usd))
funds_per_company = na.omit(funds_per_company)

print(qplot(company_name, total, data= funds_per_company, geom = "histogram", stat = "identity", xlab="Company Names", ylab="Funds Received"
            , main="Funding for Each Company"))
```

```python
%%R

head(funds_per_company)
```

```python
%%R -w 1000 -h 3000

funds_per_company_and_category = ddply(investments, .(company_name,company_category_code), summarize, total = sum(raised_amount_usd))
funds_per_company_and_category = na.omit(funds_per_company_and_category)

print(qplot(company_category_code, total, data= funds_per_company_and_category, geom = "histogram", stat = "identity"
           , xlab="Company Category", ylab="Funds Received",main="Funding for every Company Category"))
```

```python
%%R

funds_per_company_and_category1 = ddply(investments, .(company_name,company_category_code), total = sum(raised_amount_usd))
funds_per_company_and_category1 = na.omit(funds_per_company_and_category1)

print(qplot(company_name, total, data = funds_per_company_and_category1, color=company_category_code, geom = "line")+facet_wrap(~company_category_code))
      
      
```
