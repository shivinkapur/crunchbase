#### Trying model for random company (Tumblr)

```python
%%R -h 600 -w 600
library(splines)
library(data.table)
#Now lets take a random company (say Tumblr)

Tumblr=subset(web.company.funds.new, web.company.funds.new$company_name %in% c("Tumblr"))
Tumblr=data.table(Tumblr)
Tumblr$funded_year=sort(Tumblr$funded_year)
print(Tumblr)
print(plot(Tumblr$funded_year, Tumblr$total))

yearlims=range(Tumblr$funded_year)
year.grid=seq(from=yearlims[1], to = yearlims[2])

fit1 = lm(Tumblr$total ~ bs(Tumblr$funded_year,knots = c(2, 4, 6)), data = Tumblr)
pred1 = predict(fit1, newdata = list(funded_year = year.grid), se = T)
lines(year.grid, pred1$fit1, lwd = 2)
print(max(abs(pred1$y-Tumblr$total)))
print(pred1)
print(pred1$fit1)
```

```python
%%R

print(plot(Tumblr$funded_year, Tumblr$total))

fit2 = lm(Tumblr$total ~ ns(Tumblr$funded_year, df = 4), data = Tumblr)
pred2 = predict(fit2, newdata = list(funded_year = year.grid), se = T)
lines(year.grid,  pred2$fit2, col = "red", lwd = 2)
print( max(abs(pred2$y-Tumblr$total)) )
```

### Using different Distributions

```python
%%R

library(MASS)

evaluateParams= function(n){

print(n)
x=subset(web.company.funds.new, web.company.funds.new$company_name %in% c(n))

freq=x$total
years=x$funded_year
yearsFreq=rep(years, freq)

x.param.weibull = fitdistr( yearsFreq, "weibull" )
x.param.gamma = fitdistr(yearsFreq, "gamma" )
x.param.lognormal = fitdistr(yearsFreq, "lognormal")

    
print(x.param.weibull)
print(x.param.weibull$loglik)
print(x.param.gamma)
print(x.param.gamma$loglik)
print(x.param.lognormal)
print(x.param.lognormal$loglik)

}

companiesToUse= c("Tumblr", "Answers Corporation",  "Groupon", "Glassdoor", "Eventbrite",  "Business Insider",  "TaskRabbit ")
companiesToUse=as.character(companiesToUse)
for(i in 1:length(companiesToUse)){
    evaluateParams(companiesToUse[i])
}
```

#### Choosing Log Normal Distribution

```python
%%R


finalEval= function(n){

x=subset(web.company.funds.new, web.company.funds.new$company_name %in% c(n))

freq=x$total
years=x$funded_year
yearsFreq=rep(years, freq)
    
x.param.lognormal = fitdistr(yearsFreq, "lognormal")

meanlog=x.param.lognormal$estimate[1]
sdlog= x.param.lognormal$estimate[2]
loglik= x.param.lognormal$loglik

l=c(n, meanlog, sdlog, loglik)
return(l)
}

namesToUse= unique(web.company.funds.new$company_name)
namesToUse=as.character(namesToUse)
print(head(namesToUse))
print(mode(namesToUse))

N <- 1e4
fullListParams <- data.frame(name=rep("", N), meanlog=rep(NA, N), sdlog=rep(NA, N), loglik=rep(NA, N),
                 stringsAsFactors=FALSE) 

for(i in 1:length(namesToUse)){
    fullListParams[i,]= finalEval(namesToUse[i])
}

cat()
```

```python
%%R

print(head(fullListParams))
print(dim(fullListParams))
print(length(unique(fullListParams)))

fullListParams$meanlog=as.numeric(fullListParams$meanlog)
fullListParams$sdlog=as.numeric(fullListParams$sdlog)
fullListParams$loglik=as.numeric(fullListParams$loglik)
print(summary(fullListParams))

hist(fullListParams$loglik)

#Computing the 95% confidence interval of the log likelihood measure 
library(Rmisc)
loglik.confint= CI(fullListParams$loglik, ci=0.95)
print(loglik.confint)
```
