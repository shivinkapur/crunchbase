```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import brewer2mpl
import pylab
```

```python
investments = pd.read_csv('Data/investments.csv')
investments=investments.dropna(how='all')
investments.head()
```

```python
investments=investments.replace(np.nan, '0')
investments['raised_amount_usd']=investments['raised_amount_usd'].str.replace(',','')
investments.raised_amount_usd = investments.raised_amount_usd.astype('float64')

investments['funded_at']= pd.to_datetime(investments['funded_at'])
investments['funded_year']=investments.funded_year.astype('int')


print(investments.dtypes)
investments
```

```python
print(investments.columns.values.tolist())
```

```python
# Get "Set2" colors from ColorBrewer (all colorbrewer scales: http://bl.ocks.org/mbostock/5577023)
set2 = brewer2mpl.get_map('Set2', 'qualitative', 8).mpl_colors
set3 = brewer2mpl.get_map('Set3', 'qualitative', 8).mpl_colors

category_name_counts=investments.groupby('company_category_code').count()[['company_name']]

category_name_counts.plot(kind='bar',color=[set3[x] for x in range(4)], figsize=(20, 6))
```

```python
category_name_counts=investments.groupby('company_country_code').count()[['company_name']]
#category_name_counts.sort_index(by='company_country_code', ascending=False)[:1000]
category_name_counts.plot(kind='bar',color=[set3[x] for x in range(8)], figsize=(20, 6))
```

```python
category_name_counts=investments.groupby('company_region').count()[['company_name']]

category_name_counts.plot(kind='bar',color=[set3[x] for x in range(8)], figsize=(20, 6))
```

```python
category_name_counts=investments.groupby('investor_name').count()[['company_name']]

category_name_counts.plot(kind='bar',color=[set3[x] for x in range(8)], figsize=(20, 6))
```

```python
total_funds_per_year=investments.pivot_table('raised_amount_usd',rows='funded_year', aggfunc=sum)
total_funds_per_year
plot(total_funds_per_year)
```

```python
total_funds_per_quarter=investments.pivot_table('raised_amount_usd',rows='funded_quarter', aggfunc=sum)
total_funds_per_quarter
plot(total_funds_per_quarter)
```

```python
funded_year=investments.funded_year.replace(0,nan).dropna()
funded_year.hist(bins=30, color='r', figsize=(8, 6), edgecolor='w', grid=False)
```
